```

# Task 1: Unit Conversion Program

def convert_units():
    """
    Converts between different units of measurement.

    Conversion Types:
    1. Length (meters <-> feet)
    2. Weight (kg <-> pounds)
    3. Volume (liters <-> gallons)

    Returns:
        float: Converted value
    """

    print("Choose conversion type:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    try:
        choice = int(input("Enter choice (1/2/3): "))
        value = float(input("Enter value to convert: "))

        if choice == 1:
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            sub_choice = int(input("Enter option: "))

            if sub_choice == 1:
                result = value * 3.28084
                print(f"{value} meters = {result} feet")
                return result
            elif sub_choice == 2:
                result = value / 3.28084
                print(f"{value} feet = {result} meters")
                return result

        elif choice == 2:
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            sub_choice = int(input("Enter option: "))

            if sub_choice == 1:
                result = value * 2.20462
                print(f"{value} kg = {result} lbs")
                return result
            elif sub_choice == 2:
                result = value / 2.20462
                print(f"{value} lbs = {result} kg")
                return result

        elif choice == 3:
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            sub_choice = int(input("Enter option: "))

            if sub_choice == 1:
                result = value * 0.264172
                print(f"{value} L = {result} gal")
                return result
            elif sub_choice == 2:
                result = value / 0.264172
                print(f"{value} gal = {result} L")
                return result
        else:
            print("Invalid conversion type.")
            return None

    except ValueError:
        print("Invalid input. Please enter numeric values.")
        return None

# Run
convert_units()
```

```

# Task 2: Mathematical Operations


def calculate_sum(numbers):
    """Returns the sum of a list of numbers."""
    return sum(numbers)

def calculate_average(numbers):
    """Returns the average of a list of numbers."""
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    """Returns the maximum value in a list."""
    return max(numbers)

def find_minimum(numbers):
    """Returns the minimum value in a list."""
    return min(numbers)


def math_operations():
    try:
        nums = input("Enter numbers separated by spaces: ")
        numbers = [float(x) for x in nums.split()]

        if len(numbers) == 0:
            print("List is empty.")
            return

        print("Choose operation:")
        print("1. Sum")
        print("2. Average")
        print("3. Maximum")
        print("4. Minimum")

        choice = int(input("Enter choice: "))

        if choice == 1:
            print("Sum =", calculate_sum(numbers))
        elif choice == 2:
            print("Average =", calculate_average(numbers))
        elif choice == 3:
            print("Maximum =", find_maximum(numbers))
        elif choice == 4:
            print("Minimum =", find_minimum(numbers))
        else:
            print("Invalid choice.")

    except ValueError:
        print("Invalid input.")

# Run
math_operations()
```

```
# 4.2 Exercise on List Manipulation
# 1. Extract Every Other Element
def extract_every_other(lst):
    """Returns every other element starting from first."""
    return lst[::2]


# 2. Get Sublist (inclusive)
def get_sublist(lst, start, end):
    """Returns sublist from start to end (inclusive)."""
    return lst[start:end+1]


# 3. Reverse List
def reverse_list(lst):
    """Returns reversed list using slicing."""
    return lst[::-1]


# 4. Remove First and Last
def remove_first_last(lst):
    """Removes first and last elements."""
    return lst[1:-1]


# 5. Get First n Elements
def get_first_n(lst, n):
    """Returns first n elements."""
    return lst[:n]


# 6. Get Last n Elements
def get_last_n(lst, n):
    """Returns last n elements."""
    return lst[-n:]


# 7. Reverse Skip
def reverse_skip(lst):
    """Returns every second element starting from second-last, backwards."""
    return lst[-2::-2]
```

```
#  4.3 Exercise on Nested List
# 1. Flatten Nested List (Simple)
def flatten(lst):
    """Flattens one level nested list."""
    result = []
    for item in lst:
        for element in item:
            result.append(element)
    return result
# 2. Access Nested Element
def access_nested_element(lst, indices):
    """
    Returns element from nested list using list of indices.
    """
    element = lst
    for index in indices:
        element = element[index]
    return element
# 3. Sum Nested (Any Depth)
def sum_nested(lst):
    """Returns sum of all elements in nested list."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total
# 4. Remove Element from Nested List
def remove_element(lst, elem):
    """Removes all occurrences of elem from nested list."""
    new_list = []
    for item in lst:
        if isinstance(item, list):
            new_list.append(remove_element(item, elem))
        elif item != elem:
            new_list.append(item)
    return new_list
# 5. Find Maximum in Nested List
def find_max(lst):
    """Finds maximum element in nested list."""
    max_value = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_value = max(max_value, find_max(item))
        else:
            max_value = max(max_value, item)
    return max_value
# 6. Count Occurrences in Nested List
def count_occurrences(lst, elem):
    """Counts occurrences of elem in nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count
# 7. Deep Flatten (Any Depth)
def deep_flatten(lst):
    """Flattens deeply nested list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result
# 8. Average of Nested List
def average_nested(lst):
    """Returns average of nested list elements."""
    flat_list = deep_flatten(lst)
    return sum(flat_list) / len(flat_list)
```

```
#Matrix operations
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
print("A * A^-1:\n", np.dot(A, A_inv))

# 2. Prove AB != BA
print("\nAB:\n", np.dot(A,B))
print("\nBA:\n", np.dot(B,A))

# 3. Prove (AB)^T = B^T A^T
left = np.dot(A,B).T
right = np.dot(B.T, A.T)
print("\n(AB)^T:\n", left)
print("\nB^T A^T:\n", right)
```
